```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from celluloid import Camera
```

```python
def f(x, t):
    return (x + t) ** 2 * 0.5 + 2 * (np.sin(10 * (x - t)))

#Initial plot
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 50)
line, = ax.plot(x, f(x, 0), color='blue', label='time 0')
ax.set_xlabel('x')
ax.set_ylabel('f(x, t)')
ax.legend()

time_template = 'Time: %.2f'
time_text = ax.text(0.8, 0.9, '', transform=ax.transAxes, fontsize=12, color='red')

# Function to update the plot for each frame
def update(t):
    line.set_ydata(f(x, t))
    time_text.set_text(time_template % t)
    return line, time_text

# Create the animation
ani = FuncAnimation(fig, update, frames=np.linspace(0, 20, 200), blit=True)

plt.show()

ani.save('animation.mp4', writer='ffmpeg')
```
