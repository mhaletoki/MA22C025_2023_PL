```python
import numpy as np
from scipy.sparse import diags
import matplotlib.pyplot as plt
import math
```

```python
def finite_difference_matrix(n):
    D = np.zeros((n - 1, n)) # Initialize an (n-1) x n matrix with zeros
    for i in range(n - 1):
      for j in range(n):
        if i==j:
          D[i,j]=-1
        elif i==j-1:
          D[i,j]=1
        else:
          D[i,j]=0
    return D

```

```python
D=finite_difference_matrix(1000)
print(D)
```

```python
n=1000
h=2*math.pi/n
x = np.linspace(0, 2*math.pi,n)
f_x = np.sin(x)
df_x = np.dot(D, f_x)
df_x = np.dot(D, f_x)/h #dividing by h scales the finite difference approximations to match the units of the derivative

```

```python
plt.plot(x[1:],df_x,label ='approx derivative of sin(x)')
plt.plot(x,f_x,label = 'sin(x)')
plt.legend()
plt.show()
```

```python
n=1000
h=2*math.pi/n
x = np.linspace(0, 2*math.pi,n)
f_x = x**2+ 2* np.sin(10*x)
df_x = np.dot(D, f_x)
df_x = np.dot(D, f_x)/h #dividing by h scales the finite difference approximations to match the units of the derivative
```

```python
plt.plot(x[1:],df_x,label ='approx derivative of sin(x)')
plt.plot(x,f_x,label = 'x**2+2*sin(10*x)')
plt.legend()
plt.show()
```
