```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

**Plot
the solution for r(0)=5 and x(0)=2 in  0 ≤ t ≤ 70. ***

```python
def func(y, t, alpha, beta, delta, gamma): # Define the ODE system
    r, x = y
    return [alpha * r - beta * r * x,delta * r * x - gamma * x]
```

```python
 # try out some parameter values
alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.1
```

```python
# Initial conditions
r0 = 5
x0 = 2
t = np.linspace(0, 70, 1000)
```

```python

# Solve the ODEs
solution = odeint(func, [r0, x0], t, args=(alpha, beta, delta, gamma))
```

```python
# Extract solutions
r, x = solution.T

```

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(t, r, label='Rabbits (r)',color="red")
ax.plot(t, x, label='Foxes (x)',color="black")
ax.set_xlabel('Time (t)')
ax.set_ylabel('Population')
ax.set_title('Predator Prey model for Rabbits and Foxes')
ax.grid()
ax.legend()
plt.show()




```

**Prepare a figure (phase portrait) with phase plots of 10 solutions, one each for randomly
chosen initial values r(0) and x(0) between 1 and 9.**

```python
n=10 #number of initial solutions
```

```python
initial_conditions = np.random.uniform(1, 9, (n, 2)) #generate 10 random numbers for (r,x)
```

```python
# Create a colormap for different trajectories
colors = plt.cm.jet(np.linspace(0, 1, n))
```

```python
plt.figure(figsize=(7,7))
for i in range(n):
    sol = odeint(func, initial_conditions[i], t, args=(alpha, beta, delta, gamma))
    plt.plot(sol[:, 1], sol[:, 0], label=f'Trajectory {i+1}', color=colors[i])

```

**Solving the equilibria**

```python
import sympy as sp
r, x = sp.symbols('r x')
alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.1
drdt = alpha * r - beta * r * x
dxdt = delta * r * x - gamma * x
equilibria = sp.solve([drdt, dxdt], (r, x)) #solving equilibria
print("Equilibria:")
for eq in equilibria:
    print(f"({eq[0]}, {eq[1]})")
```

**Plotting the equilibria**

```python
equilibria = [(0.0, 0.0), (1.00000000000000, 2.75000000000000)]
for eq in equilibria:
    plt.scatter(eq[0], eq[1], color='red', marker='o', s=50, label=f'Equilibria{eq}')

plt.xlabel('Foxes (x)')
plt.ylabel('Rabbits (r)')
plt.title('Phase Portrait of the Predator-Prey Model')
plt.legend()
plt.grid()
plt.show()
```

**Marking the equilibria in the phase portrait**
