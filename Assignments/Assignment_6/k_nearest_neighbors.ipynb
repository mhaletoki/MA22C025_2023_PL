```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def K_nearest_neighbors(points, k):
  pairwise_distances = np.sqrt(np.sum((points[:, np.newaxis] - points) ** 2, axis=2))
  np.fill_diagonal(pairwise_distances, np.inf)
  k_nearest_indices = np.argpartition(pairwise_distances, k, axis=1)[:, :k]
  k_nearest_neighbors = points[k_nearest_indices]
  return k_nearest_neighbors
```

```python
P = np.array([[0,0], [0.2, 0.22], [0.1, -0.1],
[1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
[0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23],
[0.8, 0.6]])
k = 3
```

```python
k_nearest_neighbors = K_nearest_neighbors(P, k)

# Plot the points
plt.scatter(P[:, 0], P[:, 1], c='b', marker='o', label='Points')

# Plot arrows from each point to its k-nearest neighbors
for i in range(len(P)):
    for j in range(k):
        plt.arrow(P[i, 0], P[i, 1], k_nearest_neighbors[i, j, 0] - P[i, 0], k_nearest_neighbors[i, j, 1] - P[i, 1], head_width=0.01, head_length=0.1, fc='y', ec='g')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(f'K Nearest Neighbors (k={k})')
plt.legend()
plt.grid(True)
plt.show()





```
