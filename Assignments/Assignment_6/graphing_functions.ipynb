**Graph of sin(x) for x in [0, 10]**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.arange(0,10,0.05)
f_x=np.sin(x)

```

```python
plt.plot(x, f_x, label='sin(x)')
plt.title('Graph of sin(x) for x in [0, 10]')
plt.legend()
plt.show()
```

**Graph of cos(sqrt(x^2 + y^2)) for (x, y) in [-5, 5] × [-5, 5]**

```python
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting
```

```python
x=np.linspace(-5,5,1200)
y=np.linspace(-5,5,1200)
```

```python
x1,y1=np.meshgrid(x, y)
```

```python
z = np.cos(np.sqrt(x1**2 + y1**2))
```

```python
fig = plt.figure(figsize=(10, 8))
myplot = plt.axes(projection='3d')

# Plot the 3D surface
surface = myplot.plot_surface(x1, y1, z, cmap='viridis')

myplot.set_xlabel('x')
myplot.set_ylabel('y')
myplot.set_zlabel('cos(sqrt(x^2 + y^2))')
myplot.set_title('Graph of cos(sqrt(x^2 + y^2)) for (x, y) in [-5, 5] × [-5, 5]')

plt.show()



```
