```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x)
def f(x):
    if x < 0:
        return x * np.sin(x**2)
    else:
        return np.cos(x)
```

```python
# Create an array of x values
x = np.linspace(-5, 5, 1000001)
```

```python
# Calculate the corresponding y values using np.vectorize
y = np.vectorize(f)(x)
```

```python
#Plot the function
plt.figure(figsize=(5,5))
plt.plot(x[x < 0], y[x < 0], label='f(x) for x < 0', color='blue')
plt.plot(x[x >= 0], y[x >= 0], label='f(x) for x >= 0', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) for -5 <= x <= 5')
plt.grid(True)
plt.legend()
plt.show()
```
